1. Из признака **"Образование и ВУЗ"** в формате: **<Уровень образования год выпуска ВУЗ специальность...>** выделим признак **"Образование"**, имеющий четыре категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

```python
# текущий уровень образования находится по первым двум словам значения в столбце "Образование и ВУЗ"
def parse_education(input, sep = ' '):

    if input is None or len(input) == 0:
        return np.nan

    arr = input.split(' ')
    stop = arr.index('образование')
    education = ' '.join(input.split(sep)[:stop])

    return education

# добавим категориальный признак "Образование"
df['Образование'] = df['Образование и ВУЗ'].apply(parse_education).astype('category')


# проверим уникальность категорий
print("Уникальные категории в столбце \"Образование\":\n")
print(
    ', '.join(list(df['Образование'].unique()))
)

# отобразим статистику по категориям образования
print("\n\nКоличество соискателей в зависимости от образования:")
display(
    df['Образование'].value_counts()
)

# удалим признак "Образование и ВУЗ"
df.drop('Образование и ВУЗ', axis = 1, inplace = True);
```

2. Из признака **"Пол, возраст"** в формате **<Пол , возраст , дата рождения >** выделим два отдельных признака **"Пол"** и **"Возраст"**:

* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.


```python
# разобьем значение из столбца "Пол, возраст" по разделителю ' , '
# преобразуем первый элемент результата к буквенному значению, второй - к числовому


# выделим пол из строки
def parse_gender(input, sep = ' , '):

    if input is None or len(input) == 0:
        return np.nan

    arr = input.split(sep)

    # пол - первый элемент массива
    # нас интересует только первая буква
    gender = arr[0][0].upper()

    return gender


# выделим возраст из строки
def parse_age(input, sep = ' , '):

    if input is None or len(input) == 0:
        return np.nan

    arr = input.split(sep)

    # возраст - второй элемент массива
    # нас интересует только число
    age = arr[1].strip(' ').split(' ')[0]

    return int(age)


# добавим новые признаки
df['Пол'] = df['Пол, возраст'].apply(parse_gender).astype("category")
df['Возраст'] = df['Пол, возраст'].apply(parse_age)


# процентное соотношение соискателей по половому признаку
print("Процентное соотношение соискателей по половому признаку:")
display(
    round(df['Пол'].value_counts(normalize=True) * 100, 2)
)


# средний возраст соискателей
print("Средний возраст соискателей:")
display(
    round(df['Возраст'].mean(), 1)
)

# удалим признак "Пол, возраст"
df.drop('Пол, возраст', axis = 1, inplace = True);
```

3. Из признака **"Опыт работы"** в формате: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>** выделим общий опыт работы соискателя в месяцах - "Опыт работы (месяц)"

Условия:
* В данном признаке есть пропуски NaN - оставляем их как есть.
* В данном признаке есть скрытые пропуски: для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN.
* Нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* Необходимо учитывать, что в исходных данных опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…


```python
# разобьем строку по пробелам - первые шесть элементов содержат данные об опыте работы;
# если элемент массива содержит ключевое слово - следующий за ним элемент содержит необходимые данные
def parse_experience(input):

    nan_ = 'Не указано'
    year = ['год', 'лет']
    month = 'месяц'

    if input == nan_ or pd.isnull(input):
        return None

    arr = input.split(' ')[:6]

    months = 0
    for i, val in enumerate(arr):
        if month in val:
            months += int(arr[i-1])
            continue

        if year[0] in val or year[1] in val:
            months += int(arr[i-1]) * 12
            continue

    return months


# добавим новый признак
df['Опыт работы (месяц)'] = df["Опыт работы"].apply(parse_experience)


print("Медианный опыт работы в месяцах:")
display(
    df['Опыт работы (месяц)'].median()
)


# удалим признак "Опыт работы"
df.drop('Опыт работы', axis = 1, inplace = True);
```

4. Из признака "Город, переезд, командировки" в формате **<Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>** (в скобках указаны необязательные параметры строки) выделим признаки: **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**

Условия:
* Признак **"Город"** должен содержать 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник", остальные обозначаются как "другие".
* Признаки **"Готовность к переезду"** и **"Готовность к командировкам"** должны иметь два возможных варианта: True или False


```python
# разобьем строку по разделителю ' , ' и выделим необходимые признаки

# выделим город из первого элемента массива
def parse_city(input, sep = ' , '):

    # заведем константу, чтобы остальным городам не было так обидно
    other_awesome_cities = 'другое'

    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород',
    'Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону',
    'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']

    main_cities = ['Москва', 'Санкт-Петербург']

    if input is None or len(input) == 0:
        return np.nan

    arr = input.split(sep)

    city = arr[0]

    if city in million_cities:
        return 'город-миллионник'
    if city in main_cities:
        return city

    return other_awesome_cities


# выделим готовность к переезду/коммандировке из строки
def parse_relocation(input, unready_mark, relocation_mark):

    if input is None or len(input) == 0:
        return np.nan

    if len(unready_mark) == 0:
        raise ValueError('unready_mark')
    if len(relocation_mark) == 0:
        raise ValueError('relocation_mark')

    if relocation_mark in input and unready_mark not in input:
        return True

    return False


# добавим новые атрибуты
df['Город'] = df['Город, переезд, командировки'].apply(parse_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(parse_relocation, args = ('не готов', 'перее'))
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(parse_relocation, args = ('не готов', 'командировк'))


# сколько процентов соискателей живут в Санкт-Петербурге
print(
    'Процент соискателей из Санкт-Петербурга:\n',
    round(df[(df['Город']=='Санкт-Петербург')].shape[0] / df.shape[0] * 100)
)


# Сколько процентов соискателей готовы одновременно и к переездам, и к командировкам?
readiness_mask = (df['Готовность к переезду'] == True) & \
                 (df['Готовность к командировкам'] == True)
print(
    'Процент соискателей, готовых к переездам и коммандировкам:\n',
    round(df[readiness_mask].shape[0] / df.shape[0] * 100)
)


# удалим признак "Город, переезд, командировки"
df.drop('Город, переезд, командировки', axis = 1, inplace = True);
```
